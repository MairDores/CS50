````python
#Divisors

num = int(input('Write a number to divide: '))

list = range(1, num+1)

divisorsList = []

for number in list:
  if num % number == 0:
    divisorsList.append(number)

print(divisorsList)
````

````python
list1 = [1, 2, 23, 4, 25, 6, 7, 8, 9, 10]

list2 = [1, 1, 3, 4, 16, 8, 11, 23, 4]

commonList = []

for a in list1:
  if a in list2:
    if a in commonList:
      continue
    else:
      commonList.append(a)

print(commonList)
````

````python
# Conditionals - For loops

list = ['geeks', 'for', 'geeks\n']
for i in list:
  print(i)

tup = ('G33ks', '4', 'G33ks\n')
for i in tup:
  print(i)

str_a = ('abcdefg, beyoncee\n')
for i in str_a:
  print(i)

set1 = {10, 20, 30}
for x in set1:
  print(x)

print('The more you know')
````

````python
#List Ends

list1 = [1, 43, 5, 6, 11, 34, 22]
list2 = [3, 4, 5, 22, 5, 1, 43]

common_list = []

def find_common(list1, list2):
  for i in list1:
    if i in list2:
      common_list.append(i)

find_common(list1, list2)
common_list.sort()
print (common_list)
````
